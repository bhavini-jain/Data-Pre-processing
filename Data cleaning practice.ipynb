```python
missing_values_per=df_gdrive.isnull().sum()/df_gdrive.shape[0]*100
```

```python
missing_values_gre_20=missing_values_per[missing_values_per > 20].keys()
missing_values_gre_20
```

```python
df2_drop_clm=df_gdrive.drop(columns=missing_values_gre_20)
df2_drop_clm.shape
```

```python
df_num=df2_drop_clm.select_dtypes(include=['int64','float64'])
df_num.shape
```

```python
df_num.isnull().sum()
```

```python
num_var_miss=['LotFrontage','MasVnrArea','GarageYrBlt']
df_num[num_var_miss][df_num[num_var_miss].isnull().any(axis=1)]
```

```python
df_gdrive['LotConfig'].unique()
```

```python
df_gdrive[df_gdrive.loc[:,'LotConfig']=="Inside"]['LotFrontage'].replace
(np.nan,df_gdrive[df_gdrive.loc[:,'LotConfig']=="Inside"]['LotFrontage'].mean())
```

```python
df_copy=df_num.copy()
for var_class in df_gdrive['LotConfig'].unique():
    df_copy.update(df_gdrive[df_gdrive.loc[:,'LotConfig']==var_class]['LotFrontage'].replace
(np.nan,df_gdrive[df_gdrive.loc[:,'LotConfig']==var_class]['LotFrontage'].mean()))
```

```python
df_num.isnull().sum()
```

```python
df_copy=df_gdrive.copy()
num_var=['LotFrontage','MasVnrArea','GarageYrBlt']
cat_var=['LotConfig','MasVnrType','GarageType']
for num_v,cat_v in zip(num_var,cat_var):
    for var_class in df_gdrive[cat_v].unique():
        df_copy.update(df_gdrive[df_gdrive.loc[:,cat_v]==var_class][num_v].replace
(np.nan,df_gdrive[df_gdrive.loc[:,cat_v]==var_class][num_v].mean()))
```
