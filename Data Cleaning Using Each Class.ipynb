```python
df_copy=df_gdrive.copy()
num_var=['LotFrontage','MasVnrArea','GarageYrBlt']
cat_var=['LotConfig','MasVnrType','GarageType']
for num_v,cat_v in zip(num_var,cat_var):
    for var_class in df_gdrive[cat_v].unique():
        df_copy.update(df_gdrive[df_gdrive.loc[:,cat_v]==var_class][num_v].replace
(np.nan,df_gdrive[df_gdrive.loc[:,cat_v]==var_class][num_v].mean()))
```

```python
df_copy=df_gdrive.copy()
num_var=['LotFrontage','MasVnrArea','GarageYrBlt']
cat_var=['LotConfig','Exterior2nd','KitchenQual']
for num_v,cat_v in zip(num_var,cat_var):
    for var_class in df_gdrive[cat_v].unique():
        df_copy.update(df_gdrive[df_gdrive.loc[:,cat_v]==var_class][num_v].replace
(np.nan,df_gdrive[df_gdrive.loc[:,cat_v]==var_class][num_v].mean()))
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i,var in enumerate(num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df_gdrive[var],bins=20,hist=False,kde_kws={'linewidth':8,'color':'red'},label="Original")
    sns.distplot(df_copy[var],bins=20,hist=False,kde_kws={'linewidth':5,'color':'green'},label="Class")
    plt.legend()
    
```

```python
df_copy_median=df_gdrive.copy()
num_var=['LotFrontage','MasVnrArea','GarageYrBlt']
cat_var=['LotConfig','Exterior2nd','KitchenQual']
for num_v,cat_v in zip(num_var,cat_var):
    for var_class in df_gdrive[cat_v].unique():
        df_copy.update(df_gdrive[df_gdrive.loc[:,cat_v]==var_class][num_v].replace
(np.nan,df_gdrive[df_gdrive.loc[:,cat_v]==var_class][num_v].median()))
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i,var in enumerate(num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df_gdrive[var],bins=20,hist=False,kde_kws={'linewidth':8,'color':'red'},label="Original")
    sns.distplot(df_copy[var],bins=20,hist=False,kde_kws={'linewidth':5,'color':'green'},label="Mean")
    sns.distplot(df_copy_median[var],bins=20,hist=False,kde_kws={'linewidth':3,'color':'black'},label="Median")
    plt.legend()
    
```
