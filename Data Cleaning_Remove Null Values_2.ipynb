```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_gdrive=pd.read_csv('https://drive.google.com/uc?export=download&id=1bw4NPmQXDApAd78dRIi0MUWJqiGp5XYH')
```

```python
df_gdrive
```

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

```python
df_gdrive.info()
```

```python
missing_values_per=df_gdrive.isnull().sum()/df_gdrive.shape[0]*100
```

```python
missing_values_gre_20=missing_values_per[missing_values_per > 20].keys()
missing_values_gre_20
```

```python
df2_drop_clm=df_gdrive.drop(columns=missing_values_gre_20)
df2_drop_clm.shape
```

```python
df_num=df2_drop_clm.select_dtypes(include=['int64','float64'])
df_num.head()
```

```python
plt.figure(figsize=(30,30))
sns.heatmap(df_num.isnull())
```

```python
df_num[df_num.isnull().any(axis=1)]
```

```python
df_num.isnull().sum()
```

```python
missing_num_var=[var for var in df_num.columns if df_num[var].isnull().sum()>0]
missing_num_var
```

```python
plt.figure(figsize=(10,10))
for i,var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df_num[var],bins=20,kde_kws={'linewidth':5,'color':'red'})
```

```python
df_num.mean()
```

```python
df2_num_mean=df_num.fillna(df_num.mean())
df2_num_mean.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i,var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df_num[var],bins=20,kde_kws={'linewidth':8,'color':'red'},label='Original')
    sns.distplot(df2_num_mean[var],bins=20,kde_kws={'linewidth':5,'color':'green'},label='mean')
```

```python
df2_num_median=df_num.fillna(df_num.median())
df2_num_median.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i,var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df_num[var],bins=20,hist=False,kde_kws={'linewidth':8,'color':'red'},label="Original",)
    sns.distplot(df2_num_mean[var],bins=20,hist=False,kde_kws={'linewidth':5,'color':'green'},label="Mean",)
    sns.distplot(df2_num_median[var],bins=20,hist=False,kde_kws={'linewidth':3,'color':'k'},label="Median",)
```

```python
for i, var in enumerate(missing_num_var):
    plt.figure(figsize=(10,10))
    plt.subplot(3,1,1)
    sns.boxplot(df_num[var])
    plt.subplot(3,1,2)
    sns.boxplot(df2_num_mean[var])
    plt.subplot(3,1,3)
    sns.boxplot(df2_num_median[var])
```
